### 1) a) i) Identify the differences between a graph and a tree.





To state the differences between graphs and trees, we need first to define what are Graphs and Trees: both are data structures that model the relationship between entities for computers. They are essentially a visual representation of a problem in a way that a machine can understand, prioritise and do calculations in a way that solves problems efficiently. While all trees are graphs, not all graphs are trees.

To differenciate we have to think of their overall structure.

1) Trees have a root node, where all the other nodes originate from, while graphs do not have such a thing - they can start or end in different nodes, depending on convention. 

2) Graphs can have cycles, while trees don't. Cycles are loops that trees should not have in order to be considered trees.

3) Graphs can be connected or disconnected, while trees can only have connected nodes.

4) On trees there is only one path between two nodes, while on graphs you can/might have multiple paths.

5) For graphs the number of edges is not defined, while on trees the number will be determined by the number of nodes (n) - 1.

### 1) a) ii) Explain in detail how the graph is an abstraction of the problem.

When we say that a graph is an abstraction of the problem we mean that it takes the elements present in a real-life problem, translate it in a way that it can be read by humans without adding unnecessary details and model it in a comprehensive way. In computer science graphs are considered an abstract data type that illustrate a problem by establishing nodes and relationships between those nodes (edges).

As an example, we have subway stations in NYC. We can model those by setting nodes that represent each station and add edges that have values according to the distance/or average time the train takes between stations. With those informations we would use algorithms such as Dijkstra's to find the shortest path between stations.

In the graph presented as the problem, we have that it is being used to show the possible ways of solving a puzzle. To make it more clear, here is a list in detail of why graphs are abstraction of this problem:

1) Entities (nodes): in the example, the nodes can be a representation of each stage of the puzzle solving.

2) Relationships: the edges can represent the moves done towards the goal - that is solving the puzzle.

3) Initial node: It is the starting stage of the puzzle.

4) Destination node: the solved picture of the puzzle.
