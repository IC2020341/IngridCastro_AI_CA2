### a) i) Identify the differences between a graph and a tree.





To state the differences between graphs and trees, we need first to define what are Graphs and Trees: both are data structures that model the relationship between entities for computers. They are essentially a visual representation of a problem in a way that a machine can understand, prioritise and do calculations in a way that solves problems efficiently. While all trees are graphs, not all graphs are trees.

To differenciate we have to think of their overall structure.

1) Trees have a root node, where all the other nodes originate from, while graphs do not have such a thing - they can start or end in different nodes, depending on convention. 

2) Graphs can have cycles, while trees don't. Cycles are loops that trees should not have in order to be considered trees.

3) Graphs can be connected or disconnected, while trees can only have connected nodes.

4) On trees there is only one path between two nodes, while on graphs you can/might have multiple paths.

5) For graphs the number of edges is not defined, while on trees the number will be determined by the number of nodes (n) - 1.

### b) Demonstrate how Dijkstra’s algorithm would find the shortest path to the solution in Fig.1 through diagrams and written explanation of each stage.

Dijkstra's  algorithm consists in finding the shortest path between nodes in a weighted graph. In this case, we want to find the shortest path between node 'A' and 'S'. The algorithm does the process of:

1) It starts with the starting node, meaning, 'A';

2) It looks through its connections to see which are still not visited (at the first round all will be unvisited and undiscovered).

3) It sums up the value of 'A' to the value of the edge that connects 'A' to the undiscovered node.

4) The sum is put in the table and the 'A' node is accounted as 'Visited'.

5) From the discovered values, we pick the smallest and repeat it all from step 2. This is done in loops until all the nodes are visited and all the values are updated to the smallest path.

OBS: The smallest path keeps getting updated everytime a new visited node has a smaller path. If it doesn't have anything to be added to the table/ the value is the same as previous, the node is not updated.

## Dijkstra's algorithm 

For this code application of Dijkstra's algorithm I started thinking on what was the best approach. When thinking of the whole process shown above in 1.b. we can see that there is the principle of a priority queue, so I came across at multiple ways to implemented the code, and those were all cited on the references, but I have found that the best approach (and clearer code wise) was the one presented in class and also available in 'iqcode.com' (ANDYP, 2021) (also referenced below).

But for it to work and give us not only the value, but also the path from 'A' to each node, I needed to do some changes and also make it so the code counts the time it takes for the whole algorithm to run and generate the answer in a way that we can compare to the performance of the A*.

```python
## original code to be altered

import heapq


def calculate_distances(graph, starting_vertex):
    distances = {vertex: float('infinity') for vertex in graph}
    distances[starting_vertex] = 0

    pq = [(0, starting_vertex)]
    while len(pq) > 0:
        current_distance, current_vertex = heapq.heappop(pq)

        # Nodes can get added to the priority queue multiple times. We only
        # process a vertex the first time we remove it from the priority queue.
        if current_distance > distances[current_vertex]:
            continue

        for neighbor, weight in graph[current_vertex].items():
            distance = current_distance + weight

            # Only consider this new path if it's better than any path we've
            # already found.
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return distances
```
